```python
import zipfile
import os

# Extract the nested zip
with zipfile.ZipFile("bank_data/bank-additional.zip", "r") as zip_ref:
    zip_ref.extractall("bank_data")

# Check what's inside now
os.listdir("bank_data")
```

```python
import pandas as pd

df = pd.read_csv("bank_data/bank-additional/bank-additional-full.csv", sep=";")
df
```
